```python
import os
from datetime import datetime

# diretorio = "/home/felipe/code/teste_particionamento/dados/sp_micro"
for city in ["sp","df","cwb","rj"]:
    print(city)
    diretorio = f"/home/felipe/dados_vm/{city}/"
    arquivos = os.listdir(diretorio)
    arquivos_sorted = sorted(arquivos, key=lambda x: int(x[:10]))
    arquivos_com_diferenca = []

    # Contador para a quantidade de arquivos com diferença maior que 1 minuto
    quantidade = 0

    print(f"Total Arquivos: {len(arquivos_sorted)}")
    for i in range(1, len(arquivos_sorted)):
        arquivo_atual = arquivos_sorted[i][:10]
        arquivo_anterior = arquivos_sorted[i - 1][:10]
        timestamp_atual = datetime.fromtimestamp(int(arquivo_atual))
        timestamp_anterior = datetime.fromtimestamp(int(arquivo_anterior))

        diferenca = (timestamp_atual - timestamp_anterior).total_seconds()
        if diferenca > 120:  
            quantidade += 1
            arquivos_com_diferenca.append([str(timestamp_atual),str(timestamp_anterior)])

    with open(f"{city}.txt", "w") as arquivo:
        for item in arquivos_com_diferenca:
            arquivo.write(f"{item}\n")

    print("Quantidade de arquivos com diferença 1 minuto:", quantidade)
    print("Lista de Arquivos:", arquivos_com_diferenca)
```
